### Problem 5: Engineering Features

Now that you understand your business objective, we will build a basic model to get started.  Before we can do this, we must work to encode the data.  Using just the bank information features, prepare the features and target column for modeling with appropriate encoding and transformations.

````python
bank_features = ['age', 'job', 'marital', 'education', 'default', 'housing', 'loan']

X = df[bank_features].copy()
y = df['y'].map({'no': 0, 'yes': 1})
````

````python
X_encoded = pd.get_dummies(X, drop_first = True)

X_encoded.head()
````

````python
print("Feature matrix shape:", X_encoded.shape)
print("Target distribution:\n", y.value_counts(normalize=True))
````

### Problem 6: Train/Test Split

With your data prepared, split it into a train and test set.

````python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y,
    test_size = 0.3,
    random_state = 42,
    stratify = y
)
````

````python
print("Train:", X_train.shape, " Test:", X_test.shape)
print("\nTrain target balance:\n", y_train.value_counts(normalize=True))
print("\nTest target balance:\n", y_test.value_counts(normalize=True))
````

### Problem 7: A Baseline Model

Before we build our first model, we want to establish a baseline.  What is the baseline performance that our classifier should aim to beat?

````python
from sklearn.dummy import DummyClassifier

baseline = DummyClassifier(strategy = 'most_frequent')
baseline.fit(X_train, y_train)

baseline_acc = baseline.score(X_test, y_test)
print("Baseline (always predict 'no') accuracy: ", baseline_acc)
````

The baseline performance that our classifier should aim to beat is 88.7% accuracy. If it is beloy than it is doing wrose than no model at all. 

### Problem 8: A Simple Model

Use Logistic Regression to build a basic model on your data.  

````python
from sklearn.linear_model import LogisticRegression

lgr = LogisticRegression(max_iter = 1000)
lgr.fit(X_train, y_train)
````

### Problem 9: Score the Model

What is the accuracy of your model?

````python
train_acc = lgr.score(X_train, y_train)
test_acc = lgr.score(X_test, y_test)

print("Logistic Regression train accuracy: ", train_acc)
print("Logistic Regression test accuracy: ", test_acc)
print("Baseline accuracy: ", baseline_acc)
````

### Problem 10: Model Comparisons

Now, we aim to compare the performance of the Logistic Regression model to our KNN algorithm, Decision Tree, and SVM models.  Using the default settings for each of the models, fit and score each.  Also, be sure to compare the fit time of each of the models.  Present your findings in a `DataFrame` similar to that below:

| Model | Train Time | Train Accuracy | Test Accuracy |
| ----- | ---------- | -------------  | -----------   |
|     |    |.     |.     |

````python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
import time
````

````python
lgr = LogisticRegression(max_iter = 1000)
start = time.time()
lgr.fit(X_train, y_train)
lgr_time = time.time() - start

lgr_train = lgr.score(X_train, y_train)
lgr_test = lgr.score(X_test, y_test)
print("Logistic Regression done -- time: ", lgr_time)
````

````python
knn = KNeighborsClassifier()
start = time.time()
knn.fit(X_train, y_train)
knn_time = time.time() - start

knn_train = knn.score(X_train, y_train)
knn_test = knn.score(X_test, y_test)
print("KNN done -- time: :", knn_time)
````

````python
tree = DecisionTreeClassifier()
start = time.time()
tree.fit(X_train, y_train)
tree_time = time.time() - start

tree_train = tree.score(X_train, y_train)
tree_test = tree.score(X_test, y_test)
print("Decision tree done -- time: ", tree_time)
````

````python
svm = SVC()
start = time.time()
svm.fit(X_train, y_train)
svm_time = time.time() - start

svm_train = svm.score(X_train, y_train)
svm_test = svm.score(X_test, y_test)
print("SVM done -- time: ", svm_time)
````

````python
results_df = pd.DataFrame({
    'Model':          ['Logistic Regression', 'KNN', 'Decision Tree', 'SVM'],
    'Train Time':     [lgr_time, knn_time, tree_time, svm_time],
    'Train Accuracy': [lgr_train, knn_train, tree_train, svm_train],
    'Test Accuracy':  [lgr_test, knn_test, tree_test, svm_test],
})
results_df
````

### Problem 11: Improving the Model

Now that we have some basic models on the board, we want to try to improve these.  Below, we list a few things to explore in this pursuit.


- Hyperparameter tuning and grid search.  All of our models have additional hyperparameters to tune and explore.  For example the number of neighbors in KNN or the maximum depth of a Decision Tree.  
- Adjust your performance metric

````python
from sklearn.metrics import classification_report

for name, model in [('Logistic Regression', lgr), ('KNN', knn),
                    ('Decision Tree', tree), ('SVM', svm)]:
    print(f"\n=== {name} ===")
    print(classification_report(y_test, model.predict(X_test), digits=3))
````

````python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

lgr_grid = GridSearchCV(
    LogisticRegression(max_iter=1000, class_weight='balanced'),
    {'C': [0.01, 0.1, 1, 10]},
    scoring='f1', cv=5, n_jobs=-1
)
lgr_grid.fit(X_train, y_train)
print("Best params:", lgr_grid.best_params_, "| CV F1:", lgr_grid.best_score_)
````

````python
knn_grid = GridSearchCV(
    Pipeline([('scale', StandardScaler()), ('knn', KNeighborsClassifier())]),
    {'knn__n_neighbors': [11, 21, 31]},
    scoring='f1', cv=5, n_jobs=-1
)
knn_grid.fit(X_train, y_train)
print("Best params:", knn_grid.best_params_, "| CV F1:", knn_grid.best_score_)
````

````python
tree_grid = GridSearchCV(
    DecisionTreeClassifier(class_weight='balanced', random_state=42),
    {'max_depth': [3, 5, 10], 'min_samples_leaf': [1, 5, 20]},
    scoring='f1', cv=5, n_jobs=-1
)
tree_grid.fit(X_train, y_train)
print("Best params:", tree_grid.best_params_, "| CV F1:", tree_grid.best_score_)
````

````python
svm_bal = SVC(class_weight='balanced')
svm_bal.fit(X_train, y_train)
print("Balanced SVM fitted")
````

````python
from sklearn.metrics import f1_score

tuned = {
    'Logistic Regression': lgr_grid.best_estimator_,
    'KNN':                 knn_grid.best_estimator_,
    'Decision Tree':       tree_grid.best_estimator_,
    'SVM (balanced)':      svm_bal,
}

improved_df = pd.DataFrame([
    {'Model': name,
     'Test Accuracy': m.score(X_test, y_test),
     'Test F1 (class=yes)': f1_score(y_test, m.predict(X_test))}
    for name, m in tuned.items()
])
improved_df
````

Based on only bank client features, Logistic Regression and SVM scored about 88.7% accuracy by predicting the clients responses, "no". finding zero subscribers. f1 created a model that could actually detect subscribers. After the grid search, the Decision Tree gave the best F1 with Logistic Regression and SVM being close up and coming. The accuracy fell for those a little bit because they predicted "yes" for the likely subscribers instead of going back to "no", which is a valid switch because the banks goal was to catch subscribers.
